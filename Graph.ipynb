```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

df  = pd.read_csv('/content/natural-disasters.csv')
```

```python
NumDead = df[['Country name','Year']].join(df.loc[:,df.columns.str.contains('Number of deaths from',case=False)])
NumDead['number_of_death'] = NumDead .iloc[:, 2:].sum(axis=1)
```

```python
Pvt_NumDead = pd.pivot_table(NumDead, index='Country name',columns='Year',values='number_of_death',aggfunc='sum')
Pvt_NumDead.head()
```

#**Decadal average: Annual number of deaths from disasters**


```python
def Display_Deaddisaster(name_country='World'):
  data = Pvt_NumDead.loc[name_country]
  fig = px.bar(data,y=data.values , hover_data=None)
  fig.update_traces(hovertemplate='{}: %{{y}}'.format(name_country),showlegend=True,name=name_country)
  fig.update_layout(
    title_text='<b>Decadal average: Annual number of deaths from disasters</b><br><sub></sub>',
    title_font_size = 25,
    title_font_family = "Liberation Serif",
    title_font_color="#5b5b5b",
    xaxis_title='',
    yaxis_title='',
    plot_bgcolor= 'white',
    #yaxis=dict(gridcolor='lightgray', gridwidth=0.5)
  )
  #title_text = '<b>Decadal average: Annual number of deaths from disasters</b>'
  #fig.update_layout(title_text=title_text, title_x=0.5, title_font=dict(size=18, color='red'))  # Adjust title font properties
  fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='lightgray', griddash='dot',tickfont=dict(color='gray'))
  fig.update_xaxes(tickfont=dict(color='gray'))
  fig.show()
Display_Deaddisaster('World')
```

```python
share_dead = pd.read_csv('/content/share-deaths-from-natural-disasters.csv')
Pvt_ShareDead = pd.pivot_table(share_dead, index='Entity', columns='Year',
                               values='Deaths - Exposure to forces of nature - Sex: Both - Age: All Ages (Percent)',
                               aggfunc='sum')
```

#**Injuries and displacement from disasters**

```python
Total_affect = df[['Country name','Year']].join(df.iloc[:,df.columns.str.contains('Number of total people affected by disasters',case=False)])
Pvt_TotalAff = pd.pivot_table(Total_affect, index='Country name',
                              columns='Year', values='Number of total people affected by disasters',aggfunc='sum')
```

```python
def Display_AffectTotal(Country1='United States',Country2='Haiti',Country3='Bangladesh',Country4='Japan',Country5='Fiji'):
  datas1 = Pvt_TotalAff.loc[Country1]
  datas2 = Pvt_TotalAff.loc[Country2]
  datas3 = Pvt_TotalAff.loc[Country3]
  datas4 = Pvt_TotalAff.loc[Country4]
  datas5 = Pvt_TotalAff.loc[Country5]

  Aff_graph = pd.DataFrame({Country1: datas1.values, Country2: datas2.values, Country3: datas3.values, Country4:datas4.values, Country5:datas5.values}, index=datas1.index)
  colors = ["#ee4035", "#f37736", "#9e3d64", "#7bc043", "#0392cf"]
  fig3 = px.line(Aff_graph, x=Aff_graph.index, y=Aff_graph.columns, labels={'index': 'Date', 'value': 'Total Affected'}, color_discrete_sequence=["#ee4035", "#f37736", "#9e3d64","#7bc043","#0392cf"])
  fig3.update_traces(hovertemplate='%{x}')
  for i, cols in enumerate(Aff_graph.columns):
    goscatter = go.Scatter(x=Aff_graph.index, y=Aff_graph[cols], mode='markers', marker=dict(size=6, color=colors[i]),showlegend=False)
    fig3.add_trace(goscatter)
  fig3.update_yaxes(showgrid=True, gridcolor='#dfe3ee', gridwidth=0.5, griddash='dot', tickfont=dict(color='gray'))
  fig3.update_xaxes( tickfont=dict(color='gray'))
  fig3.update_layout(
      title='Total Affected by Country',
      title_font = dict(size=25,color='#5b5b5b',family="Liberation Serif"),
      xaxis_title='',
      yaxis_title='',
      legend_title=None,
      plot_bgcolor='white'
  )

  fig3.show()
Display_AffectTotal()
```
